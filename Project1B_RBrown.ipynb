##Project Overview
*Fitscape* is a multi-biosensor device that will monitor physiological responses during physical activity and rest. The device will incorporate measurements from Spire (i.e., breath rate) and NeuLog GSR (i.e., skin conductivity). 

##Project Goals & Research Questions
Biosensor monitoring wearable devices have promise in the realm of epidemiologic research. Additionally, the ubiquity and ease with which these devices can be integrated into daily life can also help to improve and patient-practitioner relations. Traditionally, researchers and practitioners use data collected from a single time point or repeated measures over longer periods to assess individual health and wellness. However, the advent of wearable technologies has opened new avenues from which patient or participant-level data can be harnessed. One limiting factor to integrating these novel data collection tools into research is a lack of validation and replication analyses.  

The ***overall goal* of this project is to use different types of physiological signals to develop metrics of physical exertion.** We will assess the validity and reliability of the metrics when the body is engaged in different types of physical activity and restful positions. Some scenarios include sitting in a chair, laying flat, brisk walking or jogging on a treadmill, lifting weights, standing on one foot, and stretching.  In this study we propose the use of two less common measures of physical exertion: respiration rate and galvanic skin response. The **specific aims** of this research are:


*Please note that AIM 1 corresponds to Project 1 and Aim 2 is for Project 2.*


**Aim 1: To develop empirical metrics to identify periods of physical exertion that incorporate data from *Spire* and *NeuLog*.**
A.	To perform descriptive analyses of sensor data. 
B.	To assess relationships between sensor measures.
C.	To collect data under different activity circumstances and compare results. 
D.	To automate data visualization and reporting using a dashboard to summarize key findings.   

**Aim 2: To design and conduct a trial to assess the validity and reliability of Fitscape.** 
A.	To develop a short questionnaire to evaluate self-reported exertion.
B.	To compare metrics of physical exertion, from Aim 1, to established benchmarks (e.g., FEV,  ECG, etc.). 
C.	To develop a trial protocol that includes a small battery of activities and create associated materials to conduct this experiment. 
D.	To perform the experiment using a small sample of adults (n=10) ages 18-34.

This user-specific metric helps to gage exertion and compare it to existing, person-specific benchmarks. We will evaluate the performance of the metric using a combination of machine learning tools and trial design protocols.

###Measuring exertion
I will assess whether the AUC is an appropriate summary measure of exertion. Overall, I want to examine how far above baseline (i.e, sitting sedentary) the person reaches during low (i.e., walking) and high (ie., jumping jacks) activity. I also want to see how quickly they recover. 

####Change from baseline
myFitScapeExertion_diff = ΔAUC (Act – Sed)

myFitScapeExertion_rel = ΔAUC (Act /Sed)

####3-minute Recovery
myFitScapeRecovery_diff =ΔAUC (Rec – Sed)

myFitScapeRecovery_rel =ΔAUC (Rec /Sed)
